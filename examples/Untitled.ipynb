```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
# This sets reasonable defaults for font size for
# a figure that will go in a paper
sns.set_context("paper")
```

```python
plt.style.use(['seaborn-white', 'seaborn-paper'])
matplotlib.rc("font", family="Times New Roman")
```

```python
x = pd.read_csv('results_cpu.csv')
```

```python
x
```

```python
fig, axs = plt.subplots(figsize=(12, 10))
sns.set(font_scale=1.5)
sns.lineplot(x='hidden dim size', y='time per step', style='event', 
             err_style='bars', err_kws={'capthick': 2, 'markeredgewidth': 1, 'elinewidth': 2, 'capsize': 5}, 
             markers=['o', 'o'], dashes=True, data=x, ax=axs)
handles, labels = axs.get_legend_handles_labels()
labels = ['training', 'sampling']
axs.legend(handles[1:], labels, loc='upper left')
plt.show()
```
